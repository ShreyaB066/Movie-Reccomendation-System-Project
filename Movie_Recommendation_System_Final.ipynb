```python
import re
def clean_title(title):
    title = re.sub("[^a-zA-Z0-9 ]", "", title)
    return title
```

```python
movies["clean_title"] = movies["title"].apply(clean_title)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,2))

tfidf = vectorizer.fit_transform(movies["clean_title"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search(title):
    title = clean_title(title)
    query_vec = vectorizer.transform([title])
    similarity = cosine_similarity(query_vec, tfidf).flatten()
    indices = np.argpartition(similarity, -5)[-5:]
    results = movies.iloc[indices][::-1]
    return results
```

```python
import ipywidgets as widgets
from IPython.display import display

# Create a text input widget for movie titles
movie_input = widgets.Text(
    value='Toy Story',
    description='Movie Title:',
    disabled=False
)

# Create an output widget to display search results
movie_list = widgets.Output()

def on_type(data):
    with movie_list:
        movie_list.clear_output()  # Clear previous output
        title = data["new"]  # Get the new title from the input
        if len(title) > 5:  # Check if the title length is greater than 5
            display(search(title))  # Display search results

# Observe changes in the movie input widget
movie_input.observe(on_type, names='value')

# Display the input and output widgets
display(movie_input, movie_list)
```
